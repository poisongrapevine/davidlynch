```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
con = sqlite3.connect('D:\imdb_small_indexed.db')
cur = con.cursor()
```

```python
directors_query = """
SELECT * FROM crew
JOIN people ON people.person_id=crew.person_id
JOIN role_categories ON role_categories.id = crew.category
JOIN rating ON rating.title_id=crew.title_id
WHERE role_type = 'director'
"""
```

```python
cur.execute(directors_query)
data = cur.fetchall()
```

```python
data[:5]
```

```python
df = pd.DataFrame(data, columns=['title_id', 'person_id', 'category', 'person_id_2', 'name', 'born', 'died', 'id', 'role_type', 'title_id', 'rating', 'votes'])
```

```python
df.drop(['category', 'person_id_2', 'title_id', 'rating', 'votes'], axis = 1)
```

```python
mean_ratings = pd.DataFrame(df.groupby("name")["rating"].mean())
```
